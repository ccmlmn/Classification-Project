```python
encode_columns = {
    "Yes": 1,
    "No": 0
}

columns_to_encode = [
    "Attrition",
    "OverTime",
]

for column in data.columns:
    if column in columns_to_encode:
        data[column] = data[column].map(encode_columns)

data.drop(columns = ["EmployeeNumber","EmployeeCount","Over18"], inplace = True)
data.head()
```
